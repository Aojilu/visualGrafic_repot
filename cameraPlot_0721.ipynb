```python
import cv2
import os

import numpy as np
from numpy.random import rand
from numpy import uint8, float32, float64, log, pi, sin, cos, abs, sqrt

import matplotlib.pyplot as plt
```

```python
def PlotData(x,N):
    a=np.linspace(0,N,100)
    b=x
    
    plt.plot(a,b)
    
    plt.xlim(0,N+10)
    plt.ylim(0,np.max(x)+10)
```

```python
x=np.ones(100)
PlotData(x,100)
```

```python
import cv2

def save_frame_camera_key(device_num, dir_path, basename, ext='jpg', delay=1, window_name='frame'):
    cap = cv2.VideoCapture(device_num)

    if not cap.isOpened():
        return


    n = 0
    N=100
    nowChecking=False
    meanDatas=np.zeros(N)
    while True:
        ret, frame = cap.read()
        cv2.imshow(window_name, frame)
        key = cv2.waitKey(delay) & 0xFF
        if nowChecking:
            if n<N:
                meanDatas[n]=np.mean(frame)
                n+=1
            elif(n==N) :
                print('Check End')
                n+=1;
            #画像保持
        if key == ord('c'):
            nowChecking=!nowChecking
            n=0
            if(nowChecking):
                print('Check Start!!')
            else :
                print('Check End')
        elif key == ord('q'):
            #プロット処理
            #print(meanDatas)
            PlotData(meanDatas,N)
            break

    cv2.destroyWindow(window_name)


save_frame_camera_key(1, 'data/temp', 'camera_capture')
```
